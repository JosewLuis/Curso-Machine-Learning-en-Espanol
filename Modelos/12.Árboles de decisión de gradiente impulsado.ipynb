# Árboles de decisión de gradiente impulsado.

Es un método alternativo de árbol de decisión, donde durante el entrenamiento crea una serie de pequeños árboles de decisión. Cada árbol intenta corregir errores de la etapa previa.

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from adspy_shared_utilities import plot_class_regions_for_classifier_subplot
from sklearn.datasets import make_blobs

X,y=make_blobs(n_samples=300,n_features=2,centers=8,cluster_std=1.3,random_state=4)

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

fig,subaxes=plt.subplots(1,1,figsize=(6,6))

clf=GradientBoostingClassifier().fit(X_train,y_train)
title='ADGI,Dataset de make_blobs,configuracion por defecto.'
plot_class_regions_for_classifier_subplot(clf,X_train,y_train,X_test,y_test,title,subaxes)

plt.show()
```

Los parámetros clave de este modelo son:<br>
* n_estimators: Conjunto de los pequeños árboles de decisión para usar en el ensamblaje.
* learning_rate: Controla el énfasis de correlación de error de la iteración anterior.<br>

n_estimators suele ser ajustado primero, para un mejor uso de memoria y CPU durante el entrenamiento.<br>
El parámetro max_depth suele ser pequeño para más aplicaciones.
