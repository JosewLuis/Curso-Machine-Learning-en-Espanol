# Regresión Lineal de Grado Mayor que 1.

Consiste en generar nuevas características a partid de las combinaciones polinómicas de las características originales.<br>
\begin{equation*}
    x=(x_0,x_1)----->x'=(x_0,x_1,x_0^2,x_0x_1,x_1^2).
\end{equation*}
\begin{equation*}
    y=w_0x_0+w_1x_1+w_{00}x_0^2+w_{01}x_0x_1+w_{11}x_1^2+b.
\end{equation*}
El grado especifica cuantas variables participan en cada nueva característica.<br>
 $\textbf{Sigue siendo un modelo lineal}$.

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

#Definimos nuestros datos.
X=np.array([5,15,25,35,45,55]).reshape((-1, 1))
y=np.array([28,11,3,8,29,30])

plt.figure()
plt.plot(X,y,'ro')
```

Ahora, definiremos una regresión lineal de grado 1,2,3 y 5 para ver así la diferencia.

```python
from sklearn.linear_model import LinearRegression

#Grado 1.
lr=LinearRegression().fit(X,y)
plt.subplot(2,2,1)
plt.title('Grado 1',y=0.01)
plt.plot(X,y,'ro')
plt.plot(X,lr.coef_*X+lr.intercept_,'k-')

#Grado 2.
plt.subplot(2,2,2)
plt.title('Grado 2',y=0.01)
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
lr=LinearRegression().fit(X_poly,y)
plt.plot(X,y,'ro')
plt.plot(X,lr.coef_[2]*X**2+lr.coef_[1]*X+lr.intercept_,'k-')

#Grado 3:
plt.subplot(2,2,3)
plt.title('Grado 3',y=0.01)
poly=PolynomialFeatures(degree=3)
X_poly=poly.fit_transform(X)
lr=LinearRegression().fit(X_poly,y)
plt.plot(X,y,'ro')
plt.plot(X,lr.coef_[3]*X**3+lr.coef_[2]*X**2+lr.coef_[1]*X+lr.intercept_,'k-')

#Grado 5:
plt.subplot(2,2,4)
plt.title('Grado 5',y=0.01)
poly=PolynomialFeatures(degree=5)
X_poly=poly.fit_transform(X)
lr=LinearRegression().fit(X_poly,y)
plt.plot(X,y,'ro')
plt.plot(X,lr.coef_[5]*X**5+lr.coef_[4]*X**4+lr.coef_[3]*X**3+lr.coef_[2]*X**2+lr.coef_[1]*X+lr.intercept_,'k-')
```

Como se puede ver, es el mismo ejemplo que pusimos en el apartado de overfitting-underfitting.
