# Clasificador Knn.

Es un método de aprendizaje supervisado no paramétrico que estima el valor de la función de densidad de probabilidad o de la probabilidad.<br>
Estima la clase del punto situado en $X_iy_i$ dependiendo de los k puntos ya conocidos más cercanos.

$\textbf{Complejidad del modelo}$: Número de vecinos más cerccanos a considerar k,. Por defecto 5.<br>
$\textbf{Ajuste del modelo}$: Función de distancia entre los puntos de datos. Por defecto p=2 (euclídea).<br>

Usaremos como modelo la base de datos iris, la cual contiene 3 tipos de flores distintas.

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
import numpy as np
from sklearn import datasets

iris=datasets.load_iris()
iris_X=iris.data
iris_y=iris.target

np.unique(iris_y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris_X,iris_y,random_state=0)

knn=KNeighborsClassifier().fit(X_train,y_train)

print('Precision con train: {}'.format(knn.score(X_train,y_train)))
print('Precision con test: {}'.format(knn.score(X_test,y_test)))

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
```

El método Knn también puede ser usado como un regresor, veamos un ejemplo.

```python
from sklearn.datasets import make_regression

plt.figure()
plt.title('Sample regression problem with one input variable')
X,y=make_regression(n_samples=300,n_features=1,n_informative=1,bias=140.0,noise=30,random_state=0)
plt.plot(X,y,'ro')
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

knn=KNeighborsRegressor(n_neighbors=5).fit(X_train,y_train)

#Pintamos las predicciones que este modelo daria.
print(knn.predict(X_test))

print('R-squared error: {}'.format(knn.score(X_test,y_test)))
```
