# Machine Learning.

Machine Learning se define como un tipo de aprendizaje automático que automatiza modelos analíticos.<br>
Nació del reconocimiento de patrones, bajo la teoría de que las computadoras pueden aprender sin ser programadas.<br>

Como objetivo pretenden eliminar la necesidad de intuición o gran conocimiento en el análisis de datos.

## Modelos.

* Modelos probabilísticos: Basados en estadística bayesiana.
* Modelos lógicos: Árboles de decisión.
* Modelos geométricos.

## Tipos de algoritmos.

* Aprendizaje supervisado: El algoritmo produce una función que asemeja la entrada a la salida. La base de conocimiento del sistema está formada por ejemplos de etiquetados anteriores.
* Aprendizaje no supervisado: Todo el proceso de modelaje se lleva a cabo sobre un conjunto de datos formado tan sólo por entradas al sistema. Debe poder reconocer patrones y asociarles etiquetas.

## Regresión/Clasificación.

* Regresión: Tiene como objetivo predecir valores continuos.
* Clasificación: Se emplea cuando el resultado deseado es una etiqueta discreta.
