# Clasificador de Naïve-Bayes.

Se llama naïve porque asume que cada característca es independiente. Es un método de gran eficiencia prediciendo/aprendiendo.

Tipos:<br>
* Bernoulli: Característica binaria.
* Multinomial: Característica discreta.
* Gaussian: Característica continua/real.

Las dos primeras son empleadas en minería de datos.

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from adspy_shared_utilities import plot_class_regions_for_classifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=300,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,flip_y=0.1,
                               class_sep=0.5,random_state=0)

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

gaussian=GaussianNB().fit(X_train,y_train)
plot_class_regions_for_classifier(gaussian,X_train,y_train,X_test,y_test,'Clasificador de Naïve Bayes:')
```

```python
X,y=make_classification(n_samples=100,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,flip_y=0.1,
                               class_sep=0.5,random_state=0)

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

gaussian=GaussianNB().fit(X_train,y_train)
plot_class_regions_for_classifier(gaussian,X_train,y_train,X_test,y_test,'Clasificador de Naïve Bayes:')
```

```python
from sklearn.datasets import load_breast_cancer

cancer=load_breast_cancer()
X_cancer,y_cancer=cancer.data,cancer.target

X_train,X_test,y_train,y_test=train_test_split(X_cancer,y_cancer,random_state=0)

gaussian=GaussianNB().fit(X_train,y_train)

print('Breast Cancer dataset:')
print('Accuracy del clasificador GaussianNB en training: {:.2f}'.format(gaussian.score(X_train,y_train)))
print('Accuracy del clasificador GaussianNB en test: {:.2f}'.format(gaussian.score(X_test,y_test)))
```
