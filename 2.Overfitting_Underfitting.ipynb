# Overfitting.

Se denomina overfitting al sobreentrenamiento de un modelo de datos, de tal manera que se adapta demasiado a los datos de entrenamiento y, como se puede predecir, el resultado luego en fases de testeo será bajo, luego no es un caso favorable.

# Underfitting.

Se denomina underfiting al entrenamiento de un modelo de datos que como salida no modela los datos de entranamiento ni generealiza nuevos datos.
