# Clasificadores Lineales.

Nuestra pregunta para este tipo de problemas es: ¿Cómo debería separar dos grupos de ejemplos de entrenamiento con una línea recta?

Una vez trazada dicha línea seguimos la fórmula:
\begin{equation*}
   f(x,w,b)=sign(w\circ x+b)=sign(\sum w[i]x[i]+b).
\end{equation*}

```python
from utilidad import ejemplo_claslin

ejemplo_claslin()
```

Tengamos en cuenta las siguientes considereaciones:<br>
* w=[1,-1].
* b=0.

Luego, veremos cómo verificamos cómo es una clase u otra:
* x=-0.75. y=-2.25.<br>
  f([-0.75,-2.25],w,b)=sign(1*(-0.75)+(-1)*(-2.25)+0)=sign(-0.75+2.25)>0=+1.<br>
  Pertenece a la clase 2.
* x=-1.75. y=-0.25.<br>
  f([-0.75,-0.25],w,b)=sign(1*(-0.75)+(-1)*(-0.25)+0)=sign(-0.75+0.25)<0=-1.<br>
  Pertenece a la clase 1.

$\textbf{Margen del Clasificador}$: Definido como el máximo grosor de la frontera de decisión.<br>
$\textbf{Margen máx. del Clasificador}$: El clasificador lineal con el máx. margen es un Linear Support Vector Machine (LSVM).

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from adspy_shared_utilities import plot_class_regions_for_classifier_subplot
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=100,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,flip_y=0.1,
                               class_sep=0.5,random_state=0)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

fig,subaxes=plt.subplots(1,1,figsize=(7,5))
svc=SVC(kernel='linear',C=1).fit(X_train,y_train)
title='Linear SVC con C=1.'
plot_class_regions_for_classifier_subplot(svc,X_train,y_train,None,None,title,subaxes)
```

Como podemos intuir, C es el parámetro de normalización, modificándolo:
* Con valores grandes de C hay que seleccionar los datos de entrenamiento lo mejor posible, ya que cada uno es importante.<br>
* Con valores pequeños de C el modelo es más tolerante a errores pequeños de datos individuales.

Como ventajas este modelo tiene:<br>
* Simple y fácil de entrenar.
* Fácil de interpretar.
* Escala bien para datos muy grandes.<br>

Como desventajas este modelo tiene:<br>
* Para datos de poca dimensionalidad hay modelos mejores.
* Los datos pueden ser no-linealmente separables.
